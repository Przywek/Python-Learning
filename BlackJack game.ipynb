```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}

playing = True
```

```python
class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
    
    def __str__(self):
        return (self.rank + " of "+self.suit )
```

```python
class Deck:
    
    def __init__(self):
        self.deck = []  
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
    
    def __str__(self):
        deck_comp= " "
        for card in self.deck:
            deck_comp += "\n " + card.__str__()
        return "The deck has: "+ deck_comp

    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        card = self.deck.pop()
        return card
            
```

```python
class Dealer:
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
    
    def add_card(self,card):
        # card passed in form deck.ceal() --> single card (suit,rank)
        self.cards.append(card)
        self.value += values[card.rank] 
        # track aces
        if card.rank == "Ace":
            self.aces += 1
    def adjust_for_ace(self):
        while self.value >= 20 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
class Player:
    
    def __init__(self,name,chips):
        self.name = name
        self.chips = chips
        self.cards = []  
        self.value = 0   
        self.aces = 0   
        self.bet = 0
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank] 
        if card.rank == "Ace":
            self.aces += 1
    def adjust_for_ace(self):
        while self.value >= 20 and self.aces:
            self.value -= 10
            self.aces -= 1
    def win_bet(self, global_bet):
        self.chips += global_bet
    def lose_bet(self):
        self.chips -= self.bet
    def reset(self):
        self.cards = 0
        self.value = 0 
        self.aces = 0
```

```python
def take_bet(player):

    while True:
        try:
            player.bet = int(input('How many chips would {} like to bet? ').format(player.name))
        except ValueError:
            print('Sorry, a bet must be an integer!')
        else:
            if player.bet > player.chips:
                print("Sorry, your bet can't exceed",player.chips)
            else:
                break
```

```python
def hit(deck,player):
    player.add_card(deck.deal())
    player.adjust_for_ace()
def hit_or_stand(deck,player):
    global playing
    while True:
        x = input("Would you like to Hit or Stand? Enter 'h' or 's' ")
        
        if x[0].lower() == 'h':
            hit(deck,player)  # hit() function defined above
            playing = False

        elif x[0].lower() == 's':
            print("Player stands.")
            playing = False

        else:
            print("Sorry, please try again.")
            continue
        break
```

```python
def show_player(player):
    print("\n")
    print("{} Hand:".format(player.name))
    for card in player.cards:
        print(card)
def show_dealer(dealer):
    print("Dealers hand:")
    print("one card hidden!")
    print(dealer.cards[1])
def show_all(dealer):
    print("Dealers Hand:")
    for card in dealer.cards:
        print(card)
    print("\n")
```

```python
def player_busts(player):
    print("Bust {}!".format(player.name))
    player.lose_bet()
def player_wins(player, global_bet):
    print("{} wins!".format(player.name))
    player.win_bet(global_bet)

def dealer_busts(player,global_bet):
    print("{} wins! dealer busted!".format(player.name))
    player.win_bet(global_bet)
    
def dealer_wins(player):
    print("dealer wins!")
    player.lose_bet()
    
def push(x):
    if x<1:
        print ("Dealer and players tie! PUSH")
    else:
        print ("Dealer and player tie! PUSH")
```
